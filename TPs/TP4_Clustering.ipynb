## 3. Real dataset

Now apply both algorithms on `Mnist` dataset of hand written digits can be downloaded here: [`Mnist dataset`](https://www.kaggle.com/datasets/hojjatk/mnist-dataset) or import from `keras.datasets` module as follow:

```python
from keras.datasets import mnist
import matplotlib.pyplot as plt
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
import numpy as np
ID = np.random.randint(low=0, high=len(y_train), size=8)
fig, ax = plt.subplots(2,4, figsize=(12, 6))
for i in range(8):
    ax[i//4, i%4].imshow(X_train[ID[i],:,:])
    ax[i//4, i%4].set_title(f"True label: {y_train[ID[i]]}")
plt.tight_layout()
plt.show()
```
